```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("CombinedAllEvents_PET3 (1).txt", sep="	|\t|\n",header=None, engine="python")
df1.head()
```

```python
i = 0
df2 = pd.DataFrame()
#this sorts just the first 1000 lines of df1. to sort the entire df1, replace 1000 with 'len(df1)' --> iterates throughout entire df.
# I would keep it at 1000 though when you play around with the 2d plot
while i < 1000: 
    for j in range(int(df1[0].loc[i])):
        if (int(df1[2].loc[i+j]) >= 240) and (int(df1[2].loc[i+j]) <= 250):
            for k in range(i, i + int(df1[0].loc[i])):
                rows = df1.loc[k].to_frame().T
                df2 = pd.concat([df2,rows], ignore_index=True)
            break
    i += int(df1[0].loc[i])
```
